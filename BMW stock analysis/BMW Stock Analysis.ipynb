```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
sns.set_theme(style="whitegrid")
```

```python
stock = pd.read_csv("BMW_Data.csv")
```

```python
stock
```

# Data Cleaning

```python
stock["Date"]=pd.to_datetime(stock["Date"])
```

```python
stock = stock.sort_values("Date")
```

```python
stock
```

```python
stock.isnull().sum()
```

# Data Transformation

```python
stock["Year"] = stock["Date"].dt.year
stock["Month_name"] = stock["Date"].dt.month_name()
stock["Day_name"] = stock["Date"].dt.day_name()
```

```python
stock
```

```python
stock["Daily_return"]=stock["Adj_Close"].pct_change()
```

```python
stock["Price_range"]=stock["High"]-stock["Low"]
```

```python
stock
```

```python
stock["MA50"]=stock["Adj_Close"].rolling(window=50).mean()
stock["MA200"] = stock["Adj_Close"].rolling(window=200).mean()
```

```python
stock
```

```python
stock["Volatility_30d"] = stock["Daily_return"].rolling(window=30).std()*np.sqrt(252)
```

```python
stock
```

```python
stock_clean = stock.dropna().copy()
```

```python
Yearly_performance = stock_clean.groupby("Year").agg({
    "Adj_Close":"mean",
    "Volume":"mean",
    "Daily_return":"sum"
}).reset_index()
```

# Data Visualization

```python
# Price Trends & Moving Averages (Recent 10 Years)
```

```python
plt.figure(figsize=(12,6))
stock_recent = stock_clean[stock_clean["Year"]>=2014]
plt.plot(stock_recent["Date"],stock_recent["Adj_Close"],label = "Adj_Close",alpha=0.5)
plt.plot(stock_recent["Date"], stock_recent["MA50"], label="50-Day MA", color="orange")
plt.plot(stock_recent["Date"], stock_recent["MA200"], label="200-Day MA", color="red", linewidth=2)
plt.title("BMW Price Trends & Moving Averages (2014-2024)")
plt.legend()
```

```python
# Correlation Heatmap
plt.figure(figsize=(10, 8))
cols = ["Open", "High", "Low", "Close", "Adj_Close", "Volume", "Daily_return"]
sns.heatmap(stock_clean[cols].corr(), annot=True, cmap="RdBu", center=0)
plt.title("Correlation Matrix of Stock Metrics")
```

```python
# Monthly Avg Returns (Seasonality)
monthly_avg = stock_clean.groupby("Month_name")["Daily_return"].mean() * 100
plt.figure(figsize=(10, 5))
sns.barplot(x=monthly_avg.index, y=monthly_avg.values, palette="coolwarm")
plt.title("Average Daily Returns by Month (%)")
plt.xticks(ticks=range(12), labels=["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"])
plt.ylabel("Avg Return %")
```
